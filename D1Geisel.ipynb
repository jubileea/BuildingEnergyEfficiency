```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates # Importing mdates for date tick locator 
from datetime import datetime

data = pd.read_csv('GeiselLibrary.csv')
data['DateTime'] = pd.to_datetime(data['DateTime'], format='mixed') 
data96 = data.head(96) #Slice the DataFrame to include only the first 96 rows (data for 1 day; 4x24=96)
# Date is a Saturday
plt.figure(figsize=(12, 5))
plt.plot(data96['DateTime'], data96['RealPower'], label='Real Power')
plt.title('Power Consumed in Geisel Library over 1 Day')
plt.xlabel('Time')
plt.ylabel('Power Consumed [W]')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates # Importing mdates for date tick locator 
from datetime import datetime

data = pd.read_csv('GeiselLibrary.csv')
data['DateTime'] = pd.to_datetime(data['DateTime'], format='mixed')
start_date = '02/03/2020'
end_date = '02/10/2020'
data96 = data[(data['DateTime'] >= start_date) & (data['DateTime'] <= end_date)]

plt.figure(figsize=(12, 5))
plt.plot(data96['DateTime'], data96['RealPower'], label='Real Power')
plt.title('Power Consumed in Geisel Library over 1 Week (Monday-Sunday)')
plt.xlabel('Date')
plt.ylabel('Power Consumed [W]')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m/%d/%Y')) 
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates # Importing mdates for date tick locator 
from datetime import datetime

data = pd.read_csv('GeiselLibrary.csv')
data['DateTime'] = pd.to_datetime(data['DateTime'], format='mixed')
start_date = '01/01/2019'
end_date = '12/30/2019'
data96 = data[(data['DateTime'] >= start_date) & (data['DateTime'] <= end_date)]

plt.figure(figsize=(12, 5))
plt.plot(data96['DateTime'], data96['RealPower'], label='Real Power')
plt.title('Power Consumed in Geisel Library over 1 Year')
plt.xlabel('Month')
plt.ylabel('Power Consumed [W]')
plt.grid(True)
plt.legend()
ax = plt.gca()  #gca = GetCurrentAxis
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m')) # set label to month data
ax.xaxis.set_major_locator(mdates.MonthLocator())
plt.tight_layout()
plt.show()
```
